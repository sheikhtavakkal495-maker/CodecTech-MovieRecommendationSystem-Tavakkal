```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# Load dataset
movies = pd.read_csv(r"C:\Users\DELL\Desktop\movies.csv")
# Display first rows
print("Dataset Preview")
print(movies.head())
```

```python
# Fill missing values
movies["genres"] = movies["genres"].fillna("")
# Convert genres into numerical vectors
cv = CountVectorizer()
genre_matrix = cv.fit_transform(
    movies["genres"]
)
```

```python
# Calculate similarity
similarity = cosine_similarity(
    genre_matrix
)
```

```python
# Recommendation Function
def recommend(movie_name):

    movie_name = movie_name.lower()

    movie_index = movies[
        movies["title"].str.lower() == movie_name
    ].index

    if len(movie_index) == 0:
        print("Movie not found")
        return

    movie_index = movie_index[0]

    distances = list(
        enumerate(similarity[movie_index])
    )

    distances = sorted(
        distances,
        key=lambda x: x[1],
        reverse=True
    )

    print("\nRecommended Movies:\n")

    count = 0

    for movie in distances[1:6]:

        print(
            movies.iloc[movie[0]].title
        )

        count += 1
```

```python
# User Input
movie = input("Enter Movie Name: ")
recommend(movie)
```
